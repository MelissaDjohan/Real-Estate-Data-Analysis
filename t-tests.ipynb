```python
# Dependencies and Setup
%matplotlib inline
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
import scipy.stats as stats
from scipy.stats import sem

# Hide warning messages in notebook
import warnings
warnings.filterwarnings('ignore')

# File to Load
austin_load = "CSV/avg_austin"


# Read the Data
austin_data_df = pd.read_csv(austin_load)
austin_data_df.head()
```

```python
# remove the previous index
austin_data = austin_data_df[['Year Built', 'Sale Amount']]
austin_data.head()
```

```python
# File to Load
chicago_load = "CSV/avg_chicago"


# Read the Data
chicago_data_df = pd.read_csv(chicago_load)
chicago_data_df.head()
```

```python
# remove the previous index
chicago_data = chicago_data_df[['Year Built', 'Sale Amount']]
chicago_data.head()
```

```python
# merge
merge_one = pd.merge(austin_data, chicago_data, on="Year Built", how="outer")

# rename merge columns
merge_one  = merge_one.rename(columns = {"Sale Amount_x": "Austin Avg", "Sale Amount_y" : "Chicago Avg"})

merge_one.head()
```

```python
# File to Load
cupertino_load = "CSV/avg_cupertino"

# Read the Data
cupertino_data_df = pd.read_csv(cupertino_load)
cupertino_data_df.head()
```

```python
# remove the previous index
cupertino_data = cupertino_data_df[['Year Built', 'Sale Amount']]
cupertino_data.head()
```

```python
# merge
merge_two = pd.merge(merge_one, cupertino_data, on="Year Built", how="outer")

# rename merge columns
merge_two  = merge_two.rename(columns = {"Sale Amount": "Cupertino Avg"})

merge_two.head()
```

```python
# File to Load
syracuse_load = "CSV/avg_syracuse"

# Read the Data
syracuse_data_df = pd.read_csv(syracuse_load)
syracuse_data_df.head()
```

```python
# remove the previous index
syracuse_data = syracuse_data_df[['Year Built', 'Sale Amount']]
syracuse_data.head()
```

```python
# merge
merge_three = pd.merge(merge_two, syracuse_data, on="Year Built", how="outer")

# rename merge columns
merge_three  = merge_three.rename(columns = {"Sale Amount": "Syracuse Avg"})

merge_three
```

```python
# looking at the data, the best set to use for all cities would be 1970-1973
# call upon data using loc
four_cities = merge_three.iloc[[8,9,10,11],:]
four_cities
```

```python
# create dataframe
summary_table = pd.DataFrame(four_cities)

summary_table.set_index('Year Built', inplace=True)
summary_table
```

```python
austin = summary_table.iloc[:,[0]]
chicago = summary_table.iloc[:,[1]]
cupertino = summary_table.iloc[:,[2]]
syracuse = summary_table.iloc[:,[3]]
```

```python
# Create a boxplot to compare means
summary_table.plot(kind='box')
```

```python
# T-Tests
test_one = stats.ttest_ind(austin, chicago, equal_var=False)
test_two = stats.ttest_ind(austin, cupertino, equal_var=False)
test_three = stats.ttest_ind(austin, syracuse, equal_var=False)
test_four = stats.ttest_ind(chicago, cupertino, equal_var=False)
test_five = stats.ttest_ind(chicago, syracuse, equal_var=False)
test_six = stats.ttest_ind(syracuse, cupertino, equal_var=False)

print(test_one)
print(test_two)
print(test_three)
print(test_four)
print(test_five)
print(test_six)
```
