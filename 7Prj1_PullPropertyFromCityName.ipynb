```python
# # === four cities --  property/snapshot?cityname
# cities = ["Syracuse", "Chicago", "Austin", "Cupertino"] # each run only for one city due to big pagesize!!

city="Austin"    # Input1

base_url ="https://api.gateway.attomdata.com/propertyapi/v1.0.0/"
headers = {'apikey': "0c399a902e267040d6a97e3d53c94651",'Accept': "application/json"}   

query_url = f"{base_url}property/snapshot?cityname={city}&page=1&pagesize=10000"

rsp = requests.get(query_url, headers=headers)  

rsp_json = rsp.json()

City_Property_data=[]

print(datetime.datetime.now())

for v in  rsp_json['property']:

    if 'country' not in v['address'].keys():
        AddressCountry = ""
    else:
        AddressCountry = v['address']['country']

    if 'countrySubd' not in v['address'].keys():
        AddressCountrySubd = ""
    else:
        AddressCountrySubd = v['address']['countrySubd']

    if 'line1' not in v['address'].keys():
        AddressLine1 = ""
    else:
        AddressLine1 = v['address']['line1']

    if 'line2' not in v['address'].keys():
        AddressLine2 = ""
    else:
        AddressLine2 = v['address']['line2']

    if 'locality' not in v['address'].keys():
        AddressLocality = ""
    else:
        AddressLocality = v['address']['locality']

    if 'matchCode' not in v['address'].keys():
        AddressMatchCode = ""
    else:
        AddressMatchCode = v['address']['matchCode'] 

    if 'oneLine' not in v['address'].keys():
        AddressOneLine = ""
    else:
        AddressOneLine = v['address']['oneLine']  

    if 'postal1' not in v['address'].keys():
        AddressPostal1 = ""
    else:
        AddressPostal1 = v['address']['postal1'] 

    if 'postal2' not in v['address'].keys():
        AddressPostal2 = ""
    else:
        AddressPostal2 = v['address']['postal2']

    if 'postal3' not in v['address'].keys():
        AddressPostal3 = ""
    else:
        AddressPostal3 = v['address']['postal3'] 

    if 'bathstotal' not in v['building']['rooms'].keys():
        BuildingRoomsBathstotal = ""
    else:
        BuildingRoomsBathstotal = v['building']['rooms']['bathstotal']

    if 'beds' not in v['building']['rooms'].keys():
        BuildingRoomsBeds = ""
    else:
        BuildingRoomsBeds = v['building']['rooms']['beds']

    if 'universalsize' not in v['building']['size'].keys():
        BuildingSizeUniversalsize = ""
    else:
        BuildingSizeUniversalsize = v['building']['size']['universalsize']

    if 'apn' not in v['identifier'].keys():
        IdentifierApn = ""
    else:
        IdentifierApn = v['identifier']['apn']

    if 'attomId' not in v['identifier'].keys():
        IdentifierAttomId = ""
    else:
        IdentifierAttomId = v['identifier']['attomId']

    if 'fips' not in v['identifier'].keys():
        IdentifierFips = ""
    else:
        IdentifierFips = v['identifier']['fips']

    if 'obPropId' not in v['identifier'].keys():
        IdentifierObPropId = ""
    else:
        IdentifierObPropId = v['identifier']['obPropId']

    if 'accuracy' not in v['location'].keys():
        LocationAccuracy = ""
    else:
        LocationAccuracy = v['location']['accuracy']

    if 'distance' not in v['location'].keys():
        LocationDistance = ""
    else:
        LocationDistance = v['location']['distance']

    if 'elevation' not in v['location'].keys():
        LocationElevation = ""
    else:
        LocationElevation = v['location']['elevation']

    if 'geoid' not in v['location'].keys():
        LocationGeoid = ""
    else:
        LocationGeoid = v['location']['geoid'] 

    if 'latitude' not in v['location'].keys():
        LocationLatitude = ""
    else:
        LocationLatitude = v['location']['latitude']

    if 'longitude' not in v['location'].keys():
        LocationLongitude = ""
    else:
        LocationLongitude = v['location']['longitude']

    if 'lotSize1' not in v['lot'].keys():
        LotLotSize1 = ""
    else:
        LotLotSize1 = v['lot']['lotSize1']

    if 'propIndicator' not in v['summary'].keys():
        SummaryPropIndicator = ""
    else:
        SummaryPropIndicator = v['summary']['propIndicator']

    if 'propLandUse' not in v['summary'].keys():
        SummaryPropLandUse = ""
    else:
        SummaryPropLandUse = v['summary']['propLandUse']                    

    if 'propclass' not in v['summary'].keys():
        SummaryPropclass = ""
    else:
        SummaryPropclass = v['summary']['propclass'] 

    if 'proptype' not in v['summary'].keys():
        SummaryProptype = ""
    else:
        SummaryProptype = v['summary']['proptype']

    if 'yearbuilt' not in v['summary'].keys():
        SummaryYearbuilt = ""
    else:
        SummaryYearbuilt = v['summary']['yearbuilt']

    if 'lastModified' not in v['vintage'].keys():
        VintageLastModified = ""
    else:
        VintageLastModified = v['vintage']['lastModified']

    if 'pubDate' not in v['vintage'].keys():
        VintagePubDate = ""
    else:
        VintagePubDate = v['vintage']['pubDate']                

    City_Property_data.append([
        AddressCountry,
        AddressCountrySubd,
        AddressLine1,
        AddressLine2,
        AddressLocality,
        AddressMatchCode,
        AddressOneLine,
        AddressPostal1, 
        AddressPostal2,   
        AddressPostal3,  
        BuildingRoomsBathstotal,
        BuildingRoomsBeds,
        BuildingSizeUniversalsize,
        IdentifierApn,
        IdentifierAttomId,
        IdentifierFips,
        IdentifierObPropId,
        LocationAccuracy,
        LocationDistance,
        LocationElevation,
        LocationGeoid,
        LocationLatitude,
        LocationLongitude,
        LotLotSize1,
        SummaryPropIndicator,
        SummaryPropLandUse,
        SummaryPropclass,
        SummaryProptype,
        SummaryYearbuilt,
        VintageLastModified,
        VintagePubDate
    ])        

print(datetime.datetime.now())

City_Property_df = pd.DataFrame(City_Property_data)
City_Property_df.columns = [
    'AddressCountry',
    'AddressCountrySubd',
    'AddressLine1',
    'AddressLine2',
    'AddressLocality',
    'AddressMatchCode',
    'AddressOneLine',
    'AddressPostal1', 
    'AddressPostal2',   
    'AddressPostal3',            
    'BuildingRoomsBathstotal',
    'BuildingRoomsBeds',
    'BuildingSizeUniversalsize',
    'IdentifierApn',
    'IdentifierAttomId',
    'IdentifierFips',
    'IdentifierObPropId',
    'LocationAccuracy',
    'LocationDistance',
    'LocationElevation',
    'LocationGeoid',
    'LocationLatitude',
    'LocationLongitude',
    'LotLotSize1',
    'SummaryPropIndicator',
    'SummaryPropLandUse',
    'SummaryPropclass',
    'SummaryProptype',
    'SummaryPearbuilt',
    'VintageLastModified',
    'VintagePubDate'
]   


City_Property_df.to_csv(f"City_{city}_Property_data.csv",index=False) 

City_Property_df.head()
```
